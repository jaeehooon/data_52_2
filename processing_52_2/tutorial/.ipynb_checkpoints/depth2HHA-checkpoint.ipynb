```python
sample_path = random.sample(target_list, 1)[0]
comp, loc, scene, weather, date, numbering = sample_path
print(sample_path)
```

```python
label_dir = os.path.join(data_root, comp, loc, scene, weather, date, numbering, 'seg')
refine_dir = os.path.join(data_root, comp, loc, scene, weather, date, numbering, 'refine')
calib_dir = os.path.join(data_root, comp, loc, scene, weather, date, numbering, 'calibration')
```

```python
pcd_list = get_files(os.path.join(refine_dir, 'pcd'))
seg_list = get_files(os.path.join(label_dir, "segmentation"))
```

```python
sample_pcd_file = random.sample(pcd_list, 1)[0]
print(sample_pcd_file)
print(os.path.exists(sample_pcd_file))
```

- 이미지 읽기

```python
pcd_fn = os.path.splitext(os.path.basename(sample_pcd_file))[0]
sample_img_fn = ""
for (root, _, files) in os.walk(data_root):
    if len(files) > 0 and files[0] != '.DS_Store':
        for file in files:
            fn, ext = os.path.splitext(file)
            if fn == pcd_fn:
                sample_img_fn = os.path.join(root, "{}.png".format(fn))
                
print(sample_img_fn)
```

```python
sample_img = cv2.imread(sample_img_fn, cv2.IMREAD_COLOR)
```

```python
plt.figure(figsize=(15, 20))
plt.imshow(cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB))
plt.show()
```
